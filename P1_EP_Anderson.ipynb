```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


url="https://raw.githubusercontent.com/anderson-silveira/Probabilidade_Estatistica_2021.2/main/evolucaoadmit.csv"

admissoes = pd.read_csv(url,skiprows=1)


tamanho = len(admissoes)

siglas=admissoes["Sigla"]
estados=admissoes["Estado"]
meses=admissoes.columns[3:]

#################
#Calculando a média,mediana e desvio padrão de admissões de todos os estados
#################


medias=np.empty(tamanho)
medianas=np.empty(tamanho)
desvios=np.empty(tamanho)
for i in range(tamanho):
  medias[i]=np.around(admissoes.iloc[i,3:].mean(),decimals=1)
  medianas[i]=np.around(admissoes.iloc[i,3:].median(),decimals=1)
  desvios[i]=np.around(admissoes.iloc[i,3:].std(),decimals=1)



#################
#Gráficos de barras das média,mediana e desvio padrão de admissões de todos os estados
#################

plt.rcParams["figure.figsize"] = (10, 5)

#Médias
titulo="Médias de admissões por estado"
eixoX="Estados"
eixoY="Média de admissões"


plt.title(titulo)
plt.xlabel(eixoX)
plt.ylabel(eixoY)
plt.bar(siglas,medias)
plt.show()

#Medianas
titulo="Medianas de admissões por estado"
eixoX="Estados"
eixoY="Mediana de admissões"


plt.title(titulo)
plt.xlabel(eixoX)
plt.ylabel(eixoY)
plt.bar(siglas,medianas)
plt.show()

#Desvio Padrão
titulo="Desvio padrão de admissões por estado"
eixoX="Estados"
eixoY="Desvio padrão de admissões"


plt.title(titulo)
plt.xlabel(eixoX)
plt.ylabel(eixoY)
plt.bar(siglas,desvios)
plt.show()

#################
#DataFrame com as média,mediana e desvio padrão de admissões de todos os estados
#################

estados_frame={}

estados_frame["Estados"]=estados.values
estados_frame["Médias de admissões"]=medias
estados_frame["Medianas de admissões"]=medianas
estados_frame["Desvio padrão de admissões"]=desvios


estados_DataFrame = pd.DataFrame(estados_frame)
display(estados_DataFrame)


###########################################
#Gráfico de barras com as admissões para cada estado
############################################

for i in range(tamanho):
  titulo="Número de admissões por mês. Estado: "+estados[i]
  eixoX="Mês"
  eixoY="Número de de admissões"

  plt.title(titulo)
  plt.xlabel(eixoX)
  plt.ylabel(eixoY)
  plt.bar(meses,admissoes.iloc[i,3:])
  plt.show()

###########################################
#Boxplot com as admissões para cada estado
############################################

for i in range(tamanho):
  titulo="Boxplot de admissões. Estado: "+estados[i]
  eixoY="Número de de admissões"

  plt.title(titulo)
  plt.ylabel(eixoY)
  plt.boxplot(admissoes.iloc[i,3:])
  plt.show()


###########################################
#Scatterplot com as admissões os estados tomados dois a dois
############################################

for i in range (tamanho):
  for j in range(i+1,tamanho):
    titulo="Scatterplot das admissões por mês. "+estados[i]+" X "+estados[j]
    eixoX=estados[i]
    eixoY=estados[j]

    plt.title(titulo)
    plt.xlabel(eixoX)
    plt.ylabel(eixoY)
    plt.scatter(admissoes.iloc[i,3:],admissoes.iloc[j,3:])
    plt.show()



##########################
#Calculandos os quartis
##########################

q1=np.empty(tamanho)
q2=np.empty(tamanho)
q3=np.empty(tamanho)

for i in range(tamanho):
  q1[i]=admissoes.iloc[i,3:].quantile(0.25)
  q2[i]=admissoes.iloc[i,3:].quantile(0.5)
  q3[i]=admissoes.iloc[i,3:].quantile(0.75)

quartis_frame={}

quartis_frame["Estados"]=estados.values
quartis_frame["Primeiro quartil"]=q1
quartis_frame["Segundo quartil"]=q2
quartis_frame["Terceiro quartil"]=q3

quartis_DataFrame = pd.DataFrame(quartis_frame)
display(quartis_DataFrame)



##########################
#Gerando gráficos de distribuição de probabilidade para cada estado
##########################


for i in range(tamanho):
  titulo="Distribuição de probabilidade. Estado: "+estados[i]
  eixoX="Admissões"

  plt.title(titulo)
  plt.xlabel(eixoX)

  admissoes.iloc[i,3:].plot.density()
  plt.show()

#################################################
# Como a média de São Paulo é muito maior que os
# outros estados, então terão probabilidade de 
# 100% de estarem abaixo.
```
